```octave
function ret = mk_deg_mat( size, tap )
    ret = zeros( size, size );
    for i=0:size-1
      for j=0:tap-1
        pos = mod( i+j, size );
        ret( i+1, pos+1 ) = 1.0 / tap;
      end
    end
end
```

```octave
img = im2double(imread("LENA256.pgm"));
```

```octave
# 劣化作用素を生成
A = mk_deg_mat(256, 15);
```

```octave
# 画像を劣化させる
img_deg = A * img;
```

```octave
# 画像を復元する
img_rest = inv(A) * img_deg;
```

```octave
subplot(1, 2, 1);
imshow(img_deg);
subplot(1, 2, 2);
imshow(img_rest);
```

```octave
# ファイルへ書き込む
imwrite(img_deg, "P10_DEG.pgm");
imwrite(img_rest, "P10_REST.pgm");
```
