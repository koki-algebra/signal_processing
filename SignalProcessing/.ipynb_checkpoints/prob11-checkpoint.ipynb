```octave
# 画像読み込み
my_img_deg = im2double(imread("P10_DEG.pgm"));
img_deg = im2double(imread("LENA256DEG.pgm"));
```

```octave
subplot(1, 2, 1);
imshow(my_img_deg);
subplot(1, 2, 2);
imshow(img_deg);
```

```octave
function ret = mk_deg_mat( size, tap )
    ret = zeros( size, size );
    for i=0:size-1
      for j=0:tap-1
        pos = mod( i+j, size );
        ret( i+1, pos+1 ) = 1.0 / tap;
      end
    end
end
```

```octave
# 劣化作用素を生成
A = mk_deg_mat(256, 15);
```

```octave
# 画像を復元する
img_rest = inv(A) * img_deg;
```

```octave
my_img_rest = im2double(imread("P10_REST.pgm"));
```

```octave
subplot(1, 2, 1);
imshow(my_img_rest);
subplot(1, 2, 2);
imshow(img_rest);
```

```octave
# ファイルへ書き込む
imwrite(img_rest, "P11_REST.pgm");
```
